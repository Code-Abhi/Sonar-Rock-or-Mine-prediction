```python
#Importing the necessary libararies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA LOADING

```python
# Load the dataset
sonar_data = pd.read_csv("sonar data.csv")
#print the first few columns of the dataset
sonar_data.head()
```

DATA PREPROCESSING

```python
# Separate the features (X) and labels (y)
X = sonar_data.drop(columns="R", axis=1)
y = sonar_data["R"]
```

DATA SPLITTING

```python
train_X, val_X ,train_Y , Val_Y = train_test_split(X,y)
```

MODEL CREATION

```python
rock_model=LogisticRegression()
rock_model.fit(train_X,train_Y)
```

MODEL PREDICTION

```python
predictions=rock_model.predict(train_X)
training_accuracy=accuracy_score(predictions,train_Y)
print("training accuracy = ",training_accuracy)
```

```python
test_prediction=rock_model.predict(val_X)
testing_accuracy=accuracy_score(test_prediction,Val_Y)
print("testing accuracy = ",testing_accuracy)
```

MODEL TRAINED ON ALL DATA

```python
rock_mine_model=LogisticRegression()
rock_mine_model.fit(X,y)
rock_mine_predictions=rock_mine_model.predict(X)
accuracy=accuracy_score(y,rock_mine_predictions)
print("Accuracy of Rock Mine Model: ", accuracy*100,"%")
print("predictions \n ", rock_mine_predictions[:5])
```
